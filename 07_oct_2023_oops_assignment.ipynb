```python
#Problem 1: Bank Account Create a class representing a bank account with attributes like account number, account holder name, and balance. Implement methods to deposit and withdraw money from the account.
class bank:
    '''
    Bank Account Create a class representing a bank account with attributes like account number, account holder name, and balance. Implement methods to deposit and withdraw money from the account.
    '''
    
    def __init__(self, name, number, amount):
        
        self.name = name
        self.number = number
        self.amount = amount
        print("Your account created Successfully!")
    
    def deposit(self, amount):
        self.amount += amount
        print(f"{amount} rs. deposited into your account, available balance is: {self.amount}")
    
    def withdraw(self, amount):
        if( amount <= self.amount):
            print(f"{amount} rs deducted from your account" )
            self.amount -= amount
        else:
            print(f"In sufficient funds")
    
    def balance(self):
        print(f"Available balance is {self.amount}")

        
bank_account = bank("PWSkills", 100232032, 10000.00)
bank_account.deposit(2000)
bank_account.balance()
bank_account.withdraw(40000)
bank_account.deposit(20000)
bank_account.balance()
bank_account.withdraw(5000)
```

```python
#Problem 2: Employee Management Create a class representing an employee with attributes like employee ID, name, and salary. 
#Implement methods to calculate the yearly bonus and display employee details.

import logging as log


log.basicConfig(filename="./employee.log", level=log.INFO, format="%(asctime)s - %(levelname)s - %(message)s")

class employee:
    
    '''
        Employee Management Create a class representing an employee with attributes
        like employee ID, name, and salary. Implement methods to calculate the yearly bonus and 
        display employee details.
    '''
    
    def __init__(self, id, name, salary):
        '''
            to create an employee with id, name and salary
        '''
        log.info("Employee record initiating...")
        self.id = id
        self.name = name
        try:
            self.salary = float(salary)
            log.info("Employee  created successfull!")
        except ValueError as v:
            log.error("Unable to create user without salary")

    def cal_bonus(self):
        log.info("bonus process initiated  ")
        bonus_per = 0.40
        if(5000 < self.salary <= 10000):
            bonus_per = 0.30
        elif(10000 < self.salary <= 20000):
            bonus_per = 0.20
        elif(20000 < self.salary <= 30000):
            bonus_per = 0.18
        elif(30000 < self.salary <= 40000):
            bonus_per = 0.15
        elif(self.salary > 50000):
            bonus_per = 0.10
        bonus_amt = self.salary * bonus_per
        log.info(f"Your bonus percentage is {bonus_per} and amount is {bonus_amt}")
        self.salary += bonus_amt
        log.info(f"Your Salary with bonus: {self.salary}", )
    
    def details(self):
        log.info(f"Details: Name: {self.name}, Id: {self.id}, salary: {self.salary}")
        print(f"Details: Name: {self.name}, Id: {self.id}, salary: {self.salary}")
    
emp1 = employee('1', 'emp1', '10000')
emp1.cal_bonus()
emp1.details()
```

```python
# Problem 3: Vehicle Rental Create a class representing a vehicle rental system. 
# Implement methods to rent a vehicle, return a vehicle, and display available vehicles.

class VehicleRental:
    def __init__(self):
        self.available_vehicles = {}
        self.rented_vehicles = {}

    def add_vehicle(self, vehicle, quantity):
        if quantity > 0:
            if vehicle in self.available_vehicles:
                self.available_vehicles[vehicle] += quantity
            else:
                self.available_vehicles[vehicle] = quantity

    def rent_vehicle(self, vehicle, quantity):
        if vehicle in self.available_vehicles and self.available_vehicles[vehicle] >= quantity:
            self.available_vehicles[vehicle] -= quantity
            if vehicle in self.rented_vehicles:
                self.rented_vehicles[vehicle] += quantity
            else:
                self.rented_vehicles[vehicle] = quantity
            return f"{quantity} {vehicle}(s) rented successfully."
        else:
            return "Vehicle not available for rent."

    def return_vehicle(self, vehicle, quantity):
        if vehicle in self.rented_vehicles and self.rented_vehicles[vehicle] >= quantity:
            self.rented_vehicles[vehicle] -= quantity
            if vehicle in self.available_vehicles:
                self.available_vehicles[vehicle] += quantity
            else:
                self.available_vehicles[vehicle] = quantity
            return f"{quantity} {vehicle}(s) returned successfully."
        else:
            return "Invalid return request."

    def display_available_vehicles(self):
        if self.available_vehicles:
            return "Available Vehicles:\n" + "\n".join([f"{vehicle}: {quantity}" for vehicle, quantity in self.available_vehicles.items()])
        else:
            return "No vehicles available."


vr = VehicleRental()

vr.add_vehicle("Car", 5)
vr.add_vehicle("Bus", 10)

print(vr.display_available_vehicles())

print(vr.rent_vehicle("Car", 2))
print(vr.rent_vehicle("Bus", 5))

print(vr.display_available_vehicles())

print(vr.return_vehicle("Car", 1))

print(vr.display_available_vehicles())
```

```python
#Problem 4: Library Catalog Create classes representing a library and a book. 
#Implement methods to add books to the library, borrow books, and display available books.
class Book:
    def __init__(self, title, author):
        self.title = title
        self.author = author
        self.is_borrowed = False

    def __str__(self):
        return f"{self.title} by {self.author}"

class Library:
    def __init__(self):
        self.books = []

    def add_book(self, book):
        self.books.append(book)

    def borrow_book(self, book_title):
        for book in self.books:
            if book.title == book_title:
                if not book.is_borrowed:
                    book.is_borrowed = True
                    return f"You have borrowed: {book}"
                else:
                    return "This book is already borrowed."
        return "Book not found in the library."

    def display_available_books(self):
        available_books = [book for book in self.books if not book.is_borrowed]
        if available_books:
            return "Available books:\n" + "\n".join(map(str, available_books))
        else:
            return "No available books in the library."

# Create some books and a library
book1 = Book("Book 1", "Author 1")
book2 = Book("Book 2", "Author 2")
book3 = Book("Book 3", "Author 3")
library = Library()

# Add books to the library
library.add_book(book1)
library.add_book(book2)
library.add_book(book3)

# Display available books
print(library.display_available_books())

# Borrow a book
print(library.borrow_book("Book 2"))

# Display available books after borrowing
print(library.display_available_books())
```

```python
#Problem 5: Product Inventory Create classes representing a product and an inventory system. 
#Implement methods to add products to the inventory, update product quantity, and display available products.
class Product:
    def __init__(self, product_id, name, price, quantity):
        self.product_id = product_id
        self.name = name
        self.price = price
        self.quantity = quantity

    def __str__(self):
        return f"Product ID: {self.product_id}, Name: {self.name}, Price: Rs. {self.price}, Quantity: {self.quantity}"

class Inventory:
    def __init__(self):
        self.products = {}

    def add_product(self, product):
        if product.product_id in self.products:
            self.products[product.product_id].quantity += product.quantity
        else:
            self.products[product.product_id] = product

    def update_quantity(self, product_id, new_quantity):
        if product_id in self.products:
            self.products[product_id].quantity = new_quantity
        else:
            print("Product not found in the inventory.")

    def display_available_products(self):
        if self.products:
            return "Available Products:\n" + "\n".join([str(product) for product in self.products.values()])
        else:
            return "No products available in the inventory."


inventory = Inventory()

# Adding products to the inventory
product1 = Product(1, "Laptop", 800, 5)
product2 = Product(2, "Smartphone", 500, 10)
product3 = Product(3, "Smart watches", 200, 30)
inventory.add_product(product1)
inventory.add_product(product2)
inventory.add_product(product3)
# Display products
print(inventory.display_available_products())

# Update products
inventory.update_quantity(1, 3)

# Display products
print(inventory.display_available_products())
```

```python
#Problem 6: Shape Calculation Create a class representing a shape with attributes like length, width, and height. 
#Implement methods to calculate the area and perimeter of the shape.
import math
import logging as log
log.basicConfig(filename="./shapes_cal.log", level = log.INFO, format="%(asctime)s - %(levelname)s - %(message)s")

def shape_log(func):
    
    
    def log_info(*args):
        log.info(f"started method {func.__name__}, with {list(args)}")
        func(*args)
        log.info(f"Method {func.__name__}, done!")
    return log_info
        
        
class Shape:
    
    
    @shape_log
    def __init__(self, length=0, width=0, height=0):
        self.length = length
        self.width = width
        self.height = height
    
    @shape_log
    def area(self):
        return 0  
    
    @shape_log
    def perimeter(self):
        return 0  

class Rectangle(Shape):
    
    def area(self):
        return self.length * self.width
    
    def perimeter(self):
        return 2 * (self.length + self.width)

class Triangle(Shape):
    
    def area(self):
        return 0.5 * self.length * self.height
    
    def perimeter(self):
        return self.length + self.width + self.height

class Circle(Shape):
    
    def area(self):
        return math.pi * (self.length ** 2)
    
    def perimeter(self):
        return 2 * math.pi * self.length

    


rectangle = Rectangle(4, 5)
triangle = Triangle(3, 4, 5)
circle = Circle(3)


print(f"Rectangle: Area = {rectangle.area()}, Perimeter = {rectangle.perimeter()}")
print(f"Triangle: Area = {triangle.area()}, Perimeter = {triangle.perimeter()}")
print(f"Circle: Area = {circle.area()}, Perimeter = {circle.perimeter()}")

    
```

```python
#Problem 7: Student Management Create a class representing a student with attributes like student ID, name, and grades. 
#Implement methods to calculate the average grade and display student details.
import logging as log
log.basicConfig(filename="./student_grade.log", level = log.INFO, format="%(asctime)s - %(levelname)s - %(message)s")

def student_log(func):
    
    def log_info(*args):
        log.info(f"started method {func.__name__}, with {list(args)}")
        func(*args)
        log.info(f"Method {func.__name__}, done!")
    return log_info

class Student:
    
    @student_log
    def __init__(self, student_id, name):
        self.student_id = student_id
        self.name = name
        self.grades = []

    @student_log
    def add_grade(self, grade):
        self.grades.append(grade)

    
    def calculate_average_grade(self):
        if not self.grades:
            return 0
        return sum(self.grades) / len(self.grades)
    
    @student_log
    def display_details(self):
        print(f"Student Id: {self.student_id}, Name: {self.name}")
        log.info(f"Student Id: {self.student_id}, Name: {self.name}")
        if self.grades:
            log.info(f"Grades : {self.grades}, Average Grade: {self.calculate_average_grade()}")
            print("Grades:", self.grades)
            print(f"Grades:, {self.grades}, Average Grade: {self.calculate_average_grade()}")
        else:
            log.info("No grades available for this student.")
            print("No grades available for this student.")

s1 = Student("S001", "Student1")
s2 = Student("S002", "Student2")

# rades for the student
s1.add_grade(86)
s1.add_grade(60)
s1.add_grade(90)

s2.add_grade(95)
s2.add_grade(97)
s2.add_grade(98)

# Display student details
s1.display_details()
print("----------------------------")
s2.display_details()

```

```python
#Problem 8: Email Management Create a class representing an email with attributes like sender, recipient, and subject. 
#Implement methods to send an email and display email details.
import logging as log
log.basicConfig(filename="./emails.log", level = log.INFO, format="%(asctime)s - %(levelname)s - %(message)s")

def emails_log(func):

    def log_info(*args):
        log.info(f"started method {func.__name__}, with {list(args)}")
        func(*args)
        log.info(f"Method {func.__name__}, done!")
    return log_info

class Email:
    @emails_log
    def __init__(self, sender, recipient, subject):
        self.sender = sender
        self.recipient = recipient
        self.subject = subject
        self.content = ""
    
    @emails_log
    def compose(self, content):
        self.content = content
    
    @emails_log
    def send(self):
        print("Email Sent!")
    
    @emails_log
    def display_details(self):
        print("Sender:", self.sender)
        print("Recipient:", self.recipient)
        print("Subject:", self.subject)
        print("Content:")
        print(self.content)


email1 = Email("qwerty@pwskills.com", "admin@pwskills.com", "Data Science Class Invitation")
email1.compose("Dear Qwerty, \n\nYou are invited to a Data Science Class at Saturday and sunday 9 AM.")
email1.display_details()
email1.send()
```

```python
#Problem 9: Social Media Profile Create a class representing a social media profile with attributes like username and posts. 
#Implement methods to add posts, display posts, and search for posts by keyword.
import logging as log

log.basicConfig(filename="./social_media.log", level = log.INFO, format="%(asctime)s - %(levelname)s - %(message)s")

def social_media_log(func):
    def media_info_log(*args):
        log.info(f"{func.__name__} executeion started with {args}")
        func(*args)
        log.info(f"{func.__name__} executeion ended with {args}")
    return media_info_log
        

class SocialMediaProfile:
    
    @social_media_log
    def __init__(self, username):
        self.username = username
        self.posts = []

    @social_media_log
    def add_post(self, post_content):
        self.posts.append(post_content)

    @social_media_log 
    def display_posts(self):
        print(f"Posts for {self.username}:")
        for i, post in enumerate(self.posts, start=1):
            print(f"Post {i}: {post}")

    @social_media_log
    def search_posts(self, keyword):
        matching_posts = [post for post in self.posts if keyword in post]
        if matching_posts:
            print(f"Posts containing '{keyword}':")
            for i, post in enumerate(matching_posts, start=1):
                print(f"Post {i}: {post}")
        else:
            print(f"No posts found containing '{keyword}'.")

# Create a social media profile
p = SocialMediaProfile("pwskills")

p.add_post("Learn Data science Saturday & Sunday!")
p.add_post("Just finished Python!!!")

p.display_posts()

p.search_posts("Python")
p.search_posts("Bigdata")

```

```python
#Problem 10: ToDo List Create a class representing a ToDo list with attributes like tasks and due dates. 
#Implement methods to add tasks, mark tasks as completed, and display pending tasks.
import logging as log

log.basicConfig(filename = "./todolist.log", level = log.INFO, format="%(asctime)s - %(levelname)s - %(message)s")

def todo_list_log(func):
    def todo_info_log(*args):
        log.info(f"{func.__name__} execution started with {args}")
        func(*args)
        log.info(f"{func.__name__} execution ended.")
    return todo_info_log
        

class ToDoList:
    
    @todo_list_log
    def __init__(self):
        self.tasks = []

    @todo_list_log
    def add_task(self, task, due_date):
        self.tasks.append({"task": task, "due_date": due_date, "completed": False})

    @todo_list_log
    def mark_task_as_completed(self, task):
        for t in self.tasks:
            if t["task"] == task:
                t["completed"] = True

    @todo_list_log
    def display_pending_tasks(self):
        pending_tasks = [t for t in self.tasks if not t["completed"]]
        if pending_tasks:
            print("Pending Tasks:")
            for i, task in enumerate(pending_tasks, start=1):
                print(f"{i}. Task: {task['task']}, Due Date: {task['due_date']}")
        else:
            print("No pending tasks.")

my_todo_list = ToDoList()

my_todo_list.add_task("Finish the report", "2023-09-15")
my_todo_list.add_task("Buy groceries", "2023-09-10")
my_todo_list.add_task("Complete Assignment", "2023-09-12")

my_todo_list.mark_task_as_completed("Buy groceries")

my_todo_list.display_pending_tasks()
```
